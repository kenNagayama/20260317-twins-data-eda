```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib_fontja
```

```python
fpath = "src/20220916_koga-st_5cm-original-data.xlsx"
df = pd.read_excel(fpath)
```

```python
data = df.dropna(subset=['摩耗_測定値']).copy()
data.shape
```

```python
data.plot(x='キロ程', y='摩耗_測定値', kind='line', title='摩耗_測定値 vs キロ程', xlabel='キロ程', ylabel='摩耗_測定値');
```

```python
ax = sns.histplot(data['摩耗_測定値'], bins=20, kde=False)
ax.set(xlabel='摩耗_測定値', ylabel='頻度', title='摩耗_測定値の分布')
```
